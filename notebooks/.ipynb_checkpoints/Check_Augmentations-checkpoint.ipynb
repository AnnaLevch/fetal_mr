```python
import numpy as np

import sys
sys.path.append('..')

import nibabel as nib
from matplotlib import pyplot as plt

from data_generation.augment import augment_data
```

```python
def slice_it(arr, inds):
    return arr[inds[0][0]:inds[0][1], inds[1][0]: inds[1][1], inds[2][0]:inds[2][1]]
```

```python
vol = nib.load('../../../../data/brain/FR_FSE/Pat01_Se05_Res0.7422_0.7422_Spac5/volume.nii.gz')
mask = nib.load('../../../../data/brain/FR_FSE/Pat01_Se05_Res0.7422_0.7422_Spac5/truth.nii.gz')
vol.shape
```

# Cropping Augmentation

```python
from nilearn.image.image import _crop_img_to as crop_img_to, new_img_like
from nilearn import image

def find_bounding_box(mask):
    coords = np.array(np.where(mask > 0))
    start = coords.min(axis=1)
    end = coords.max(axis=1) + 1

    assert np.sum(mask[start[0]:end[0], start[1]:end[1], start[2]:end[2]]) == np.sum(mask)
    return start, end

def cut_bounding_box(img, start, end):
    slices = [slice(s, e) for s, e in zip(start, end)]
    return crop_img_to(img, slices, copy=True)

def pad_data(data,mask, padding_start, padding_end):
    bbox_start, bbox_end = find_bounding_box(mask.get_data())

    bbox_start = np.maximum(bbox_start - padding_start, 0)
    bbox_end = np.minimum(bbox_end + padding_end, mask.shape)

    volume = cut_bounding_box(vol, bbox_start, bbox_end)
    truth = cut_bounding_box(mask, bbox_start, bbox_end)
    
    return volume, truth

def crop_start_end(min_crop, max_crop):
    x_rand = np.random.uniform(min_crop[0],max_crop[0], 2)
    y_rand = np.random.uniform(min_crop[1],max_crop[1], 2)
    z_rand = np.random.uniform(min_crop[2],max_crop[2], 2)
    padding_start = np.array([int(x_rand[0]), int(y_rand[0]), int(z_rand[0])])
    padding_end = np.array([int(x_rand[1]), int(y_rand[1]), int(z_rand[1])])
    print(padding_start)
    print(padding_end)
    return padding_start, padding_end
```

```python
min_crop = np.array([16, 16, 8])
max_crop = np.array([200, 200, 10])
padding_start, padding_end = crop_start_end(min_crop, max_crop)
volume, truth = pad_data(vol,mask, padding_start, padding_end)

volume.shape
```
